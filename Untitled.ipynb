McCulloch-Pitts

import numpy as np
import matplotlib.pyplot as plt

```python
def thresh(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
def M_Pitts(x1, x2, t, w, b):
    y_preds=[]
    for i in range(len(x)):
        y=x1[i]*w[0]+x2[i]*w[1]+b
        Y_pred=thresh(y)
        y_preds.append(Y_pred)
    return y_preds
```

```python
def plotM_Pitts(x, y, t, w, b):
   plt.scatter(x, y, c=t, cmap='coolwarm', marker='s')
   # Equation of desction boundary
   X = np.linspace(-0.2, 1.2, 100)
   Y = [-xi*(w[0]/w[1]) - b/w[1] for xi in X]
   plt.plot(X, Y, color='black', label=f'y={-w[0]/w[1]}x - {b/w[1]}')
   plt.legend()
   plt.xlabel('x')
   plt.ylabel('y')
   plt.title('McCulloch-Pitts Decision Boundary for b = {} w = [{},{}]'.format(b, w[0], w[1]))
   plt.grid(True, alpha=0.3)
   plt.figure(figsize=(4, 4)) 
```
